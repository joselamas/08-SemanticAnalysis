```python
!pip install -U llama-index-core llama-index-llms-huggingface llama-index-embeddings-huggingface transformers accelerate bitsandbytes huggingface-hub

from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, Document, VectorStoreIndex
import torch

# Definimos el nombre del modelo para usarlo en ambos lugares
model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
Settings.llm = HuggingFaceLLM(
    model_name=model_id,
    tokenizer_name=model_id, # <--- Sincronización exacta aquí
    context_window=2048,
    max_new_tokens=150,
    generate_kwargs={
        "temperature": 0.1, #indica el nivel de creatividad.
        "do_sample": True
    },
    model_kwargs={
        "load_in_4bit": True,
        "torch_dtype": torch.float16,
        "trust_remote_code": True # Asegura que el código del tokenizer sea compatible
    },
    device_map="auto",
)
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
print(" Tokenizer y Modelo sincronizados. GPU lista.")
#finalizada la preconfiguracion del modelo, ejecutamos la “prueba”
# A. Creamos el dato directamente en memoria (evita problemas de lectura de archivos)
texto_secreto = "El macho del Colibrí Esmeralda se distingue por una mancha blanca circular tras el ojo."
documento = Document(text=texto_secreto)
pregunta = "¿Qué marca física tiene el macho del Colibrí Esmeralda?"
# B. Respuesta SIN RAG
print("Consultando memoria de la IA...")
res_simple = Settings.llm.complete(pregunta)
# C. Respuesta CON RAG
print("Consultando con el dato aumentado...")
index = VectorStoreIndex.from_documents([documento])
query_engine = index.as_query_engine()
res_rag = query_engine.query(pregunta)
# RESULTADOS
print("\n" + "="*60)
print(f"1. SIN RAG (IA pura): {res_simple}")
print("-" * 60)
print(f"2. CON RAG (Dato exacto): {res_rag}")
print("="*60)
```
